```python
import pandas as pd
import numpy as np
import os
import glob
import math
import sys
from mrs_parser import MRS_Parser
from video_parser import Video_Parser
from config import mrs_root, video_root

# MRS PARSER COST
active_mode = 'cost'
parser = MRS_Parser(mrs_root, mode=active_mode)
d = parser.parse()

writer = pd.ExcelWriter('mrs_dataset_with_costs(' + active_mode + ').xlsx', engine='xlsxwriter')
for key, value in d.items():
    value.to_excel(writer, sheet_name = key)
writer.save()
writer.close()

# MRS PARSER QOS
active_mode = 'qos'
parser = MRS_Parser(mrs_root, mode=active_mode)
d = parser.parse()

writer = pd.ExcelWriter('mrs_dataset_with_costs(' + active_mode + ').xlsx', engine='xlsxwriter')
for key, value in d.items():
    value.to_excel(writer, sheet_name = key)
writer.save()
writer.close()

# VIDEO PARSER COST
active_mode = 'cost'
parser = Video_Parser(video_root, mode=active_mode)
d = parser.parse()

writer = pd.ExcelWriter('video_dataset_with_costs(' + active_mode + ').xlsx', engine='xlsxwriter')
for key, value in d.items():
    value.to_excel(writer, sheet_name = key)
writer.save()
writer.close()

# VIDEO PARSER QOS
active_mode = 'qos'
parser = Video_Parser(video_root, mode=active_mode)
d = parser.parse()

writer = pd.ExcelWriter('video_dataset_with_costs(' + active_mode + ').xlsx', engine='xlsxwriter')
for key, value in d.items():
    value.to_excel(writer, sheet_name = key)
writer.save()
writer.close()

print("Finished!")
```
